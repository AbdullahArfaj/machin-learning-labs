# Exploratory Data Analysis (EDA)
## Global Cars Dataset

This notebook performs Exploratory Data Analysis (EDA) on the Global Cars Dataset.

**Objectives:**
- Understand dataset structure
- Identify missing values
- Perform statistical analysis
- Visualize relationships between variables


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('global_cars_dataset_synthetic.csv')

# Display first rows
df.head()
```

```python
# Dataset information
df.info()
```

```python
# Summary statistics
df.describe()
```

```python
# Check missing values
df.isnull().sum()
```

## Data Visualization

```python
# Histogram of numerical features
df.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
corr = df.corr(numeric_only=True)
plt.figure()
plt.imshow(corr)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.title('Correlation Matrix')
plt.show()
```

## Observations
- Identify key patterns
- Note correlations
- Highlight important insights


## Additional Visualizations

```python
# Scatter Plot: Engine Size vs Price
plt.figure()
plt.scatter(df['Engine_CC'], df['Price_USD'])
plt.xlabel('Engine_CC')
plt.ylabel('Price_USD')
plt.title('Engine Size vs Price')
plt.show()
```

```python
# Boxplot: Price Distribution
plt.figure()
plt.boxplot(df['Price_USD'])
plt.title('Price Distribution')
plt.ylabel('Price_USD')
plt.show()
```

```python
# Bar Chart: Average Price by Brand
avg_price_brand = df.groupby('Brand')['Price_USD'].mean()
plt.figure()
avg_price_brand.plot(kind='bar')
plt.title('Average Price by Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Plot: Average Price by Year
avg_price_year = df.groupby('Manufacture_Year')['Price_USD'].mean()
plt.figure()
avg_price_year.plot()
plt.title('Average Price by Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```
