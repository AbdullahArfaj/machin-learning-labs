## Advanced Options

Beyond local editing two things are quite important: editing the notebooks in the markdown format and running Jupyter remotely.
The latter matters when we want to run the code on a faster server.
The former matters since Jupyter's native ipynb format stores a lot of auxiliary data that is
irrelevant to the content,
mostly related to how and where the code is run.
This is confusing for Git, making
reviewing contributions very difficult.
Fortunately there is an alternative---native editing in the markdown format.

### Markdown Files in Jupyter

If you wish to contribute to the content of this book, you need to modify the
source file (md file, not ipynb file) on GitHub.
Using the notedown plugin we
can modify notebooks in the md format directly in Jupyter.


First, install the notedown plugin, run the Jupyter Notebook, and load the plugin:

```
pip install d2l-notedown  # You may need to uninstall the original notedown.
jupyter notebook --NotebookApp.contents_manager_class='notedown.NotedownContentsManager'
```

You may also turn on the notedown plugin by default whenever you run the Jupyter Notebook.
First, generate a Jupyter Notebook configuration file (if it has already been generated, you can skip this step).

```
jupyter notebook --generate-config
```

Then, add the following line to the end of the Jupyter Notebook configuration file (for Linux or macOS, usually in the path `~/.jupyter/jupyter_notebook_config.py`):

```
c.NotebookApp.contents_manager_class = 'notedown.NotedownContentsManager'
```

After that, you only need to run the `jupyter notebook` command to turn on the notedown plugin by default.

### Running Jupyter Notebooks on a Remote Server

Sometimes, you may want to run Jupyter notebooks on a remote server and access it through a browser on your local computer. If Linux or macOS is installed on your local machine (Windows can also support this function through third-party software such as PuTTY), you can use port forwarding:

```
ssh myserver -L 8888:localhost:8888
```

The above string `myserver` is the address of the remote server.
Then we can use http://localhost:8888 to access the remote server `myserver` that runs Jupyter notebooks. We will detail on how to run Jupyter notebooks on AWS instances
later in this appendix.

### Timing

We can use the `ExecuteTime` plugin to time the execution of each code cell in Jupyter notebooks.
Use the following commands to install the plugin:

```
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
jupyter nbextension enable execute_time/ExecuteTime
```

## Perform Initial Feature Engineering

### Subtask:
Prepare the dataset for modeling by defining the target variable, dropping leakage and identifier columns, converting datetime columns, and engineering new time-based, price-based, and distance-based features as outlined in Lab 5.


````python
y = df['Order_Status']
X = df.drop('Order_Status', axis=1)

leakage_and_identifier_columns = [
    'Order_ID', 'User_ID', 'Restaurant_ID', 'Driver_ID',
    'Delivery_Time', 'Delivery_Duration_Minutes', 'Driver_Availability'
]
X = X.drop(columns=leakage_and_identifier_columns)

print("Shape of X after dropping columns:", X.shape)
print("First 5 rows of X after dropping columns:")
print(X.head())
````

````python
X['Order_Time'] = pd.to_datetime(X['Order_Time'])
X['hour_of_day'] = X['Order_Time'].dt.hour
X['day_of_week'] = X['Order_Time'].dt.dayofweek
X['is_weekend'] = X['day_of_week'].isin([5, 6]).astype(int) # 5 for Saturday, 6 for Sunday
X['month'] = X['Order_Time'].dt.month

X['price_per_item'] = X['Total_Price'] / X['Quantity']

print("Shape of X after adding time and price features:", X.shape)
print("First 5 rows of X with new features:")
print(X.head())
````

**Reasoning**:
I will define a Python function to compute the Haversine distance and then apply this function to calculate the `restaurant_customer_distance_km` and `driver_restaurant_distance_km` features as instructed, ensuring latitude and longitude values are converted to radians for accurate calculation.



````python
import numpy as np

def haversine(lat1, lon1, lat2, lon2):
    R = 6371

    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance

X['restaurant_customer_distance_km'] = X.apply(
    lambda row: haversine(
        row['Restaurant_Lat'], row['Restaurant_Lon'],
        row['Customer_Lat'], row['Customer_Lon']
    ), axis=1
)

X['driver_restaurant_distance_km'] = X.apply(
    lambda row: haversine(
        row['Driver_Lat'], row['Driver_Lon'],
        row['Restaurant_Lat'], row['Restaurant_Lon']
    ), axis=1
)

print("Shape of X after adding distance features:", X.shape)
print("First 5 rows of X with new features:")
print(X[['restaurant_customer_distance_km', 'driver_restaurant_distance_km']].head())
````

## Perform Initial Feature Engineering

### Subtask:
Continue preparing the dataset for modeling by engineering additional features, specifically 'is_peak_hour' and 'Item_Name_reduced', and dropping original columns that have been transformed or are no longer needed.


**Reasoning**:
I will create the 'is_peak_hour' feature by checking if 'hour_of_day' falls within the specified peak hours (18-22). Then, I will identify the top 10 most frequent 'Item_Name' values and use them to create the 'Item_Name_reduced' column, grouping the rest into 'Other'. Finally, I will drop the original columns that have been transformed or are no longer necessary.



````python
X['is_peak_hour'] = ((X['hour_of_day'] >= 18) & (X['hour_of_day'] <= 22)).astype(int)


if 'Item_Name' in X.columns:
    top_10_items = X['Item_Name'].value_counts().nlargest(10).index
    X['Item_Name_reduced'] = X['Item_Name'].apply(lambda x: x if x in top_10_items else 'Other')
else:
    print("'Item_Name' not found, assuming 'Item_Name_reduced' was already created or not needed.")

    if 'Item_Name_reduced' not in X.columns:

        pass

columns_to_drop = [
    'Order_Time',
    'Restaurant_Lat', 'Restaurant_Lon',
    'Customer_Lat', 'Customer_Lon', 'Driver_Lat', 'Driver_Lon'
]

if 'Item_Name' in X.columns:
    columns_to_drop.append('Item_Name')

existing_columns_to_drop = [col for col in columns_to_drop if col in X.columns]
X = X.drop(columns=existing_columns_to_drop)

print("Shape of X after engineering more features and dropping columns:", X.shape)
print("First 5 rows of X with new features:")
print(X.head())
````

## Encode Categorical Features

### Subtask:
Encode all categorical features in the `X` DataFrame using one-hot encoding. Identify categorical columns, apply `OneHotEncoder`, and integrate the new features into the DataFrame, dropping the original categorical columns.

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

categorical_cols = X.select_dtypes(include=['object']).columns

one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', one_hot_encoder, categorical_cols)
    ],
    remainder='passthrough'
)


X_encoded = preprocessor.fit_transform(X)


encoded_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)


remaining_cols = X.select_dtypes(exclude=['object']).columns
all_feature_names = list(encoded_feature_names) + list(remaining_cols)

X_encoded = pd.DataFrame(X_encoded, columns=all_feature_names, index=X.index)

print("Shape of X after one-hot encoding:", X_encoded.shape)
print("First 5 rows of X after one-hot encoding:")
print(X_encoded.head())
````

````python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

categorical_cols = X.select_dtypes(include=['object']).columns

one_hot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', one_hot_encoder, categorical_cols)
    ],
    remainder='passthrough'
)


X_encoded = preprocessor.fit_transform(X)


encoded_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)


remaining_cols = X.select_dtypes(exclude=['object']).columns
all_feature_names = list(encoded_feature_names) + list(remaining_cols)

X_encoded = pd.DataFrame(X_encoded, columns=all_feature_names, index=X.index)

print("Shape of X after one-hot encoding:", X_encoded.shape)
print("First 5 rows of X after one-hot encoding:")
print(X_encoded.head())
````

## Train a baseline Random Forest classifier

### Subtask:
Split the preprocessed data into training and testing sets, then train a Random Forest classifier model on the training data.


````python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)


rf_classifier = RandomForestClassifier(random_state=42)


rf_classifier.fit(X_train, y_train)

print("Random Forest Classifier trained successfully.")
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
````

## Interpret model performance and feature importance

### Subtask:
Evaluate the trained Random Forest classifier on the test set and analyze its feature importances.


````python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

y_pred = rf_classifier.predict(X_test)

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {accuracy:.4f}")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


feature_importances = rf_classifier.feature_importances_


features_df = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Importance': feature_importances
})


features_df = features_df.sort_values(by='Importance', ascending=False)


print("\nTop 15 Feature Importances:")
print(features_df.head(15))


plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=features_df.head(15))
plt.title('Top 15 Feature Importances from Random Forest Classifier')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
````
